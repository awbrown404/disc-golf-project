```python
# import dependencies 
import pandas as pd
```

```python
# set up empty variable to append data to
nc_courses = []

# set up first page scrape. performed this way because url changes after first page
url = "https://www.pdga.com/course-directory/advanced/search?field_course_location_administrative_area=NC&location_title=North%20Carolina"
table = pd.read_html(url)
```

```python
# save list as a dataframe
table_1 = pd.DataFrame(table[0])
```

```python
# set up loop for the remaining pages 
for x in range(1, 6):
    url = f"https://www.pdga.com/course-directory/advanced/search?field_course_location_administrative_area=NC&location_title=North%20Carolina&page={x}"
    table = pd.read_html(url)  
    nc_courses.append(table)    
```

```python
# save lists as dataframes
table_2 = pd.DataFrame(nc_courses[0][0])
table_3 = pd.DataFrame(nc_courses[1][0])
table_4 = pd.DataFrame(nc_courses[2][0])
table_5 = pd.DataFrame(nc_courses[3][0])
table_6 = pd.DataFrame(nc_courses[4][0])
frames = [table_1, table_2, table_3, table_4, table_5, table_6]
```

```python
# perform concat to join dataframes
nc_courses_df = pd.concat(frames, join="inner", sort=True)
nc_courses_df
```

```python
# save dataframe as csv
nc_courses_df.to_csv("../data/nc_courses_df.csv")
```
