```python
# import dependencies 
from bs4 import BeautifulSoup as bs
import requests
import pymongo
import pandas as pd
from splinter import Browser
import time
# from urllib.request import urlopen
```

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# establish MongoDB connection
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.disc_golf_db
collection = db.innovadics
```

```python
# url = 'https://www.innovadiscs.com/disc/'

# https://www.innovadiscs.com/home/disc-golf-faq/plastic-types-overview/
# this is the url to get data on the plastics
```

```python
disc_name = []
disc_url = []
disc_flight_ratings = []
disc_description = []

for x in range(1,13):
    url = f'https://www.innovadiscs.com/disc/page/{x}/'
    browser.visit(url)
    time.sleep(5)
    html = browser.html
    soup = bs(html, 'html.parser')
    disc_html = soup.find_all('div', class_ = 'entry-block')

    for disc in disc_html:
        name = disc.find('a').get_text()
        disc_name.append(name)
        url = disc.find('a')['href']
        disc_url.append(url)
        try:
            flight_rating = disc.find('span').get_text()
        except:
            flight_rating = "Not Available"
        disc_flight_ratings.append(flight_rating)
        description = disc.find('p').get_text()
        disc_description.append(description)
        
    print(f"Page {x} is complete")
```

```python
# build dataframe
columns = ["Disc Name", "Disc URL", "Disc Flight Ratings", "Disc Description"]

innova_disc_db = pd.DataFrame(columns=columns)

innova_disc_db["Disc Name"] = disc_name
innova_disc_db["Disc URL"] = disc_url
innova_disc_db["Disc Flight Ratings"]= disc_flight_ratings
innova_disc_db["Disc Description"] = disc_description
innova_disc_db
```

```python
# Speed classification
# Putt & Approach speed 1 to 3
# Mid-Distance/Fairway speed 4 to 8
# Driver speed 9 to 14
disc_classification = []
import re
for disc in disc_flight_ratings:
    split = re.split("[ | ] | | ", disc)
    if split[1] == "Available":
        disc_classification.append("Ultimate, Recreation, & Specialty")
    elif float(split[1]) > 0 and float(split[1]) < 4:
        disc_classification.append("Putt & Approach")
    elif float(split[1]) > 3 and float(split[1]) < 9:
        disc_classification.append("Mid-Distance/Fairway")
    elif float(split[1]) > 8 and float(split[1]) < 15:
        disc_classification.append("Driver")
```

```python
innova_disc_db["Disc Classification"] = disc_classification
innova_disc_db
```

```python
# post data to MongoDB
for index, row in innova_disc_db.iterrows():
    post = {
        "Disc Name" : row[0],
        "Disc URL" : row[1],
        "Disc Description" : row[2],
        "Disc Flight Ratings" : row[3],
        "Disc Classification" : row[4]
    }

    collection.insert_one(post)
```

```python
innova_disc_db.to_csv("../data/innova_disc_db.csv")
```
